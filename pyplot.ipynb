### SCATTERPLOT

##### [matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

<br>
A scatterplot is a type of data display that shows the relationship between two numerical variables. [14] The data are displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis. [15]  Scatter plotsâ€™ primary uses are to observe and show relationships between two numeric variables. The dots in a scatter plot not only report the values of individual data points, but also patterns when the data are taken as a whole. [16]
<br>


````python
# Random number generator.
rng = np.random.default_rng()
````

````python
# Creating an empty figure.
fig, axs = plt.subplots(figsize = (25,7))
fig.suptitle ("SCATTERPLOT\n", size = "xx-large")

# Data.
array = rng.random([100,100])
x = array[:, 0]
y = array[:, 1]
# Naming labels.
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")
# Creating a plot.
plt.scatter(x, y)
# Display a plot.
plt.show()
````

<br>
All correlations have two properties: strength and direction. The strength of a correlation is determined by its numerical value. The direction of the correlation is determined by whether the correlation is positive or negative. [17]

- Positive correlation: When one variable increases the other one increases too, and the opposite | Variables move in the same direction
- Negative correlation: When one variable increases the other one decreases, and the opposite | Variables move in different directions
- No correlation: There is no relationship between variables
- Strength: Closer the variables are to each other (and line of fit) stronger the relationship between them

Different relationships are shown below.
<br>
<br>

````python
# Creating a figure for 7 plots.
fig, axs = plt.subplots(1, 7, figsize = (20,4))
fig.suptitle ("CORRELATION\n", size = "xx-large")

# Creating first plot.
a = rng.choice (20, 20, replace = False, shuffle = False)
axs[0].scatter (a, a, c= "#009900")
axs[0].set( title = "PERFECT POSITIVE")

# Creating second plot.
a = rng.normal (100, size =  30)
a.sort()
b = rng.normal (10, size = 30)
b.sort()
axs[1].scatter (a, b, c= "#99ff66")
axs[1].set( title = "HIGH POSITIVE")

# Creating third plot.
a = [1.0, 2.3, 2.1, 2.8, 2.8, 2.9, 3.5, 3.2, 3.1, 3.5,
     4.0, 4.3, 4.1, 4.8, 4.8, 3.9, 4.5, 4.2, 5.1, 5.5]
b = [1.1, 1.3, 1.4, 2.8, 2.3, 2.9, 3.1, 3.2, 3.6, 3.7,
     4.1, 4.3, 4.4, 4.8, 4.3, 4.9, 5.1, 5.2, 5.6, 5.7]
axs[2].scatter (a, b, c= "#c6ff1a")
axs[2].set( title = "LOW POSITIVE")

# Creating fourth plot.
a = rng.normal (100, size =  30)
b = rng.normal (10, size = 30)
axs[3].scatter (a, b, c= "#ffbf80")
axs[3].set( title = "NONE")

# Creating fifth plot.
a = [1.0, 2.3, 2.1, 2.8, 2.8, 2.9, 3.5, 3.2, 3.1, 3.5,
     4.0, 4.3, 4.1, 4.8, 4.8, 3.9, 4.5, 4.2, 5.1, 5.5]
b = [1.1, 1.3, 1.4, 2.8, 2.3, 2.9, 3.1, 3.2, 3.6, 3.7,
     4.1, 4.3, 4.4, 4.8, 4.3, 4.9, 5.1, 5.2, 5.6, 5.7]
b = np.negative(b)
axs[4].scatter (a, b, c= "#ff9966")
axs[4].set( title = "LOW NEGATIVE")

# Creating sixth plot.
a = rng.normal (100, size =  30)
a.sort()
b = rng.normal (10, size = 30)
b = np.sort (b)[::-1]
axs[5].scatter (a, b, c= "#ff6666")
axs[5].set( title = "HIGH NEGATIVE")

# Creating seventh plot.
a = rng.choice (20, 20, replace = False, shuffle = False)
b = rng.choice (20, 20, replace = False, shuffle = False)
b = np.sort(b)[::-1]
axs[6].scatter (a, b, c= "#e60000")
axs[6].set( title = "PERFECT NEGATIVE")

plt.tight_layout()
plt.show()
````

<br>
It is a common practice to use a line of best fit when plotting a scatter plot. Line of best fit refers to a line through a scatter plot of data points that best expresses the relationship between those points. [18] Points closer to the line of best fit have a stronger relationship, as stated above.

````python
# Creating a figure for 2 plots.
fig, axs = plt.subplots(1, 2, figsize = (25,7))
fig.suptitle ("SCATTERPLOT with A LINE OF BEST FIT\n", size = "xx-large")

# Data for plotting.
x = rng.normal (100, size =  100)
x.sort()
y = rng.normal (10, size = 100)
y.sort()

# First plot.
    # Line of best fit.
theta = np.polyfit(x, y, 1)
y_line = theta[1] + theta[0] * x
axs[0].plot(x, y_line, "orange")
    # Scatter plot
axs[0].scatter(x, y)


# Data for second plot.
y1 = rng.normal (10, size = 100)

# Second plot.
    # Line of best fit.
theta1 = np.polyfit(x, y1, 1)
y1_line = theta1[1] + theta1[0] * x
axs[1].plot(x, y_line, "orange")
    # Scatter plot.
axs[1].scatter(x, y1)
plt.show()
````

### HISTOGRAM
***
##### [matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, *, data=None, **kwargs)](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

A histogram is an approximate representation of the distribution of numerical data. To construct a histogram, the first step is to "bin" (or "bucket") the range of valuesâ€”that is, divide the entire range of values into a series of intervalsâ€”and then count how many values fall into each interval. [19] Histograms are commonly used in statistics to demonstrate how many of a certain type of variable occurs within a specific range. [20] In a histogram, it is the area of the bar that indicates the frequency of occurrences for each bin. This means that the height of the bar does not necessarily indicate how many occurrences of scores there were within each individual bin. It is the product of height multiplied by the width of the bin that indicates the frequency of occurrences within that bin. [21]

````python
# Creating a figure for plotting.
fig, axs = plt.subplots( figsize = (25,7))
fig.suptitle ("HISTOGRAM", size = "xx-large")

# Plotting.
your_bins = 20
data = rng.random(100)
arr = plt.hist(data,bins=your_bins, align = "left",
               edgecolor = "red", alpha = 0.5)
    # Values over bins.
for i in range(your_bins):
    plt.text(arr[1][i],(arr[0][i])+0.5,str(arr[0][i]),
             fontsize = 11)
    plt.ylim([0, 20])
    plt.xlabel("BINS", size = 15)
    plt.ylabel("FREQUENCY\n", size = 15)
# Add text and arrows.
plt.annotate("BIN", xy=(0.15, 2.5), xytext=(0.2, 10), fontsize = 15,
             arrowprops=dict(facecolor="black", shrink=0.05))

plt.show()
````

Histograms are sometimes confused with bar charts. A histogram is used for continuous data, where the bins represent ranges of data, while a bar chart is a plot of categorical variables. Some authors recommend that bar charts have gaps between the rectangles to clarify the distinction. [22] Gaps can aslo appear in histogram, but that means there are no values in the given range.

````python
# Creating a figure for 2 plots.
fig, axs = plt.subplots(1, 2, figsize = (25,7))
fig.suptitle ("HISTOGRAM vs BAR CHART\n",
               size = "xx-large")

# First plot - histogram.
data_hist = [9,10,10,1,2,3,3,4,5,5,5,3,2,7,6,6,8] 
axs[0].hist(data_hist,  edgecolor = "black", alpha = 0.5)
    # Add text and arrows.
axs[0].annotate("NO GAPS", xy=(6.5, 1), xytext=(7.6, 2), fontsize = 15,
             arrowprops=dict(facecolor="black", shrink=0.05))
axs[0].annotate("", xy=(8.2, 1), xytext=(8, 1.9), fontsize = 15,
             arrowprops=dict(facecolor="black", shrink=0.05))


# Second plot - bar plot.
data_bar = {"A":20, "B":30, "C":10, "D":10, "E":15 }
name = list(data_bar.keys())
values = list(data_bar.values())
axs[1].bar(name, values, alpha = 0.5,
          color=["red", "blue","black", "green", "orange"])
    # Add text and arrows.
axs[1].annotate("GAPS", xy=(1.5, 10), xytext=(2.2, 18), fontsize = 15,
             arrowprops=dict(facecolor="black", shrink=0.05))
axs[1].annotate("", xy=(2.5, 10), xytext=(2.4, 17), fontsize = 15,
             arrowprops=dict(facecolor="black", shrink=0.05))

plt.show()
````

<br>

| BASIS FOR COMPARISSON | HISTOGRAM | BAR PLOT |
| :- | :- | :- |
| Meaning | Histogram refers to a graphical representation, that displays data by way of bars to show the frequency of numerical data. | Bar graph is a pictorial representation of data that uses bars to compare different categories of data. |
| Indicates | Distribution of non-discrete variables | Bar graph is a pictorial representation of data that uses bars to compare different categories of data. |
| Presents | Quantitative data | Categorical data |
| Spaces | Bars touch each other, hence there are no spaces between bars | Bars do not touch each other, hence there are spaces between bars. |
| Elements | Elements are grouped together, so that they are considered as ranges. | Elements are taken as individual entities. |
| Can bars be reordered? | No | Yes  |
| Width of bars | Need not to be same | Same | 

<br>

Most common histogram shapes are:
1. Normal Distribution
2. Bimodal Distribution
3. Right-skewed Distribution
4. Left-skewed Distribution
5. Random Distribution

````python
# Creating a figure for 5 plots.
fig, axs = plt.subplots(1, 5, figsize = (20, 5))
fig.suptitle ("MOST COMMON HISTOGRAM SHAPES\n", size = "xx-large")

# Fist plot.
a = rng.normal (10, size = 1000)
axs[0].set( title = "NORMAL")
axs[0].hist(a, edgecolor = "w", alpha = 0.5)

# Second plot.
a1 = rng.normal (3, size = 1000)
a2 = rng.normal (8, size = 1000)
a = np.concatenate([a1, a2])
axs[1].set( title = "BIMODAL")
axs[1].hist(a, bins = 20, edgecolor = "w", alpha = 0.5)

# Third plot.
a = rng.hypergeometric(10, 100, 10, 100)
axs[3].set( title = "LEFT-SKEWED")
axs[3].hist(a, bins = 4, edgecolor = "w", alpha = 0.5)

# Fourth plot.
a = rng.hypergeometric(100, 10, 10, 100)
axs[2].set( title = "RIGHT-SKEWED")
axs[2].hist(a, bins = 4, edgecolor = "w", alpha = 0.5)

# Fifth plot.
a = rng.random (100)
axs[4].set( title = "RANDOM")
axs[4].hist(a, bins = 15, edgecolor = "w", alpha = 0.5)

plt.tight_layout()
plt.show()
````

<br>
When working with histograms it is very important to choose the correct bin width. There is no rule as how wide they need to be, but choosing wrong ones might lead to wrong data interpretation.

````python
# Creating a figure for 3 plots.
fig, axs = plt.subplots(1, 3, figsize = (20, 5))
fig.suptitle ("BIN SIZE\n", size = "xx-large")

# Data.
a1 = rng.normal (3, size = 100)
a2 = rng.normal (8, size = 100)
a = np.concatenate([a1, a2])

# Fist plot.
axs[0].set( title = "TOO LARGE")
axs[0].hist(a, bins = 3, edgecolor = "w", alpha = 0.5)

# Second plot.
axs[1].set( title = "NORMAL")
axs[1].hist(a, edgecolor = "w", alpha = 0.5)

# Third plot.
axs[2].set( title = "TOO SMALL")
axs[2].hist(a, bins = 1000, alpha = 0.5)

plt.tight_layout()
plt.show()
````

If bins are too large, there is too much data under one bin. Opposite, if the bins are too small, they contain only a few (or one) values. One needs to test and see what width gives the most accurate information about the data. 

### BOX PLOT
***
##### [matplotlib.pyplot.boxplot(x, notch=None, sym=None, vert=None, whis=None, positions=None, widths=None, patch_artist=None, bootstrap=None, usermedians=None, conf_intervals=None, meanline=None, showmeans=None, showcaps=None, showbox=None, showfliers=None, boxprops=None, labels=None, flierprops=None, medianprops=None, meanprops=None, capprops=None, whiskerprops=None, manage_ticks=True, autorange=False, zorder=None, *, data=None)](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html)

Box plot is a method for graphically demonstrating the locality, spread and skewness groups of numerical data through their quartiles. [23] Box plots show the five-number summary of a set of data: including the minimum score, first (lower) quartile, median, third (upper) quartile, and maximum score. [24] It can tell you about your outliers and what their values are. The outlier is the data value that differs to a high degree from the rest of the data. Box plot can be drawn both horizontally and vertically.

Box plot data:
- **Minimum (Q0 or 0th percentile)**: the lowest data point in the data set excluding any outliers
- **Maximum (Q4 or 100th percentile)**: the highest data point in the data set excluding any outliers
- **Median (Q2 or 50th percentile)**: the middle value in the data set
- **First quartile (Q1 or 25th percentile)**: also known as the lower quartile qn(0.25), it is the median of the lower half of the dataset
- **Third quartile (Q3 or 75th percentile)**: also known as the upper quartile qn(0.75), it is the median of the upper half of the dataset [25]

````python
# Creating figure for plotting.
fig = plt.figure(figsize =(9, 7))

# Creating first axes for the figure.
ax1 = fig.add_axes([1, 1, 1, 1]) 
# Creating second axes for the figure.
ax2 = fig.add_axes([2.1, 1.5, 1, 0.5])
ax3 = fig.add_axes([2.1, 1, 1, 0.5])


# Data for first (left) plot.
x = [-7, -6, 1,2,3,4,5,5,6,7,8,9, 14, 15]
median = np.median(x)
q0 = np.percentile (x, 0)
q1 = np.percentile (x, 25)
q3 = np.percentile (x, 75)
q4 = np.percentile (x, 100)

# Plotting first (left) plot and:
    # Boxplot.
ax1.boxplot(x, vert = True)
    # Add text and lines.
ax1.text (x = 1.1, y = median, s = "Median\n", fontsize = 12) 
ax1.axhline (median, ls = "--", alpha = 0.5, color = "orange")
ax1.text (x = 1.1, y = q0+1, s = "Min\n", fontsize = 12)
ax1.axhline (q0+1, ls = "--", alpha = 0.5)
ax1.text (x = 1.1, y = q4-0.3, s = "Max\n", fontsize = 12)
ax1.axhline (q4, ls = "--", alpha = 0.5)
ax1.text (x = 1.1, y = q1, s = "Q1\n", fontsize = 12)
ax1.axhline (q1, ls = "--", alpha = 0.5)
ax1.text (x = 1.1, y = q3, s = "Q3\n", fontsize = 12)
ax1.axhline (q3, ls = "--", alpha = 0.5)
ax1.text (x = 1.01, y = q0-0.7, s = "OUTLIER\n", fontsize = 12)
    # Draw vertical line and add text.
point1 = [0.8, q3]
point2 = [0.8, q1]
x_values = [point1[0], point2[0]]
y_values = [point1[1], point2[1]]
ax1.plot(x_values, y_values)
ax1.text (x = 0.8, y = median, s = "\nINTERQUARTILE RANGE", 
          fontsize = 8, rotation = "vertical", va = "center")
ax1.set( title = "BOXPLOT")

# Plotting second and third plot (right).
    # Setting x axis limit
a, b = -1, 7
x = [1.2, 2, 2.5, 3, 3.5, 4, 5]
ax2.set_xlim([a, b])

# Boxplot (top right).
bp0 = ax2.boxplot(x, vert = False, patch_artist=True)
for box in bp0["boxes"]:
    box.set(color="black")
    # change fill color
    box.set(facecolor = "#ff9999")
# Text (percentages).
ax2.text (x = 3, y = 1.1, s = "50%", fontsize = 12, ha = "center")
ax2.text (x = 1.7, y = 1.02, s = "25%", fontsize = 12, ha = "center")
ax2.text (x = 4.3, y = 1.02, s = "25%", fontsize = 12, ha = "center")
ax2.set( title = "BOXPLOT AND PDF PLOT\n Normal Distribution")

# KDE plot (bottom right).
s = pd.Series(x)
ax3 = s.plot.kde()
ax3.set_xlim([a, b])
    # Text and lines.
ax2.axvline (1.2, ls = "--", alpha = 0.5)
ax3.axvline (1.2, ls = "--", alpha = 0.5)
ax2.axvline (5, ls = "--", alpha = 0.5)
ax3.axvline (5, ls = "--", alpha = 0.5)
ax2.axvline (2.25, ls = "--", alpha = 0.5)
ax3.axvline (2.25, ls = "--", alpha = 0.5)
ax2.axvline (3.75, ls = "--", alpha = 0.5)
ax3.axvline (3.75, ls = "--", alpha = 0.5)
    # Color fill.
kde_x, kde_y = ax3.lines[0].get_data()

ax3.fill_between(kde_x, kde_y, where=(kde_x>2.25) & (kde_x<3.75), 
                interpolate=False, color="#ff9999")
ax3.fill_between(kde_x, kde_y, where= ((kde_x>1.2) & (kde_x<2.25) | 
                                          (kde_x>3.75) & (kde_x<5)),
                interpolate=False, color="b", alpha = 0.2)
    # Text (percentages).
ax3.text (x = 3, y = 0.1, s = "50%", fontsize = 12, ha = "center")
ax3.text (x = 1.7, y = 0.1, s = "25%", fontsize = 12, ha = "center")
ax3.text (x = 4.3, y = 0.1, s = "25%", fontsize = 12, ha = "center")

plt.show()
````

Boxplot can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed. [25] When boxplot is symmetric then it means data follows a normal distribution. If boxplot is not symmetric is shows that the data is skewed - left or right.

````python
# Creating a figure for 3 plots.
fig, axs = plt.subplots(1, 3, figsize = (25,7))
fig.suptitle ("BOXPLOT SKEWNESS\n\n", size = "xx-large")

# First plot
a = [1,2,3,4,5,5,6,7,8,9]
axs[0].set( title = "SYMMETRIC")
axs[0].boxplot(a, vert = False)

# Second plot.
a = [1,1,2,2,2,3,5,6,7,8,9]
axs[1].set( title = "RIGHT-SKEWED")
axs[1].boxplot(a, vert = False)

# Third plot.
a = [1,2,3,4,5,6,7,7,8,9,9,9]
axs[2].set( title = "LEFT-SKEWED")
axs[2].boxplot(a, vert = False)

plt.show()
````

````python
# Creating figure.
fig, axs = plt.subplots(1, 4, figsize = (25,7))
fig.suptitle ("FEATURE RELATIONSHIP", size = "xx-large")
z = np.arange(150)

# Plot 1.
    # Scatterplot.
axs[0].scatter (x = df["sepal_length"], y = df["petal_length"], 
                c = z, cmap = 'seismic')
axs[0].set( title = "SEPAL LENGTH vs PETAL LENGTH",
       xlabel = "SEPAL LENGTH",
       ylabel = "PETAL LENGTH")
    # Line to separate setosa.
axs[0].axline((4, 3), (7, 2))
    # Text.
axs[0].text (x = 6, y = 1, s = "SETOSA", fontsize = 10)
axs[0].text (x = 4.3, y = 5, s = "VIRGINICA\n & \nVERSICOLOR", fontsize = 10)

# Plot 2.
x = df["sepal_length"]
y = df["petal_length"]
    # Scatterplot.
axs[1].scatter (x, y)
axs[1].set( title = "SEPAL LENGTH vs PETAL LENGTH",
       xlabel = "SEPAL LENGTH",
       ylabel = "PETAL LENGTH")
    # Line of best fit.
theta = np.polyfit(x, y, 1)
y_line = theta[1] + theta[0] * x
axs[1].plot(x, y_line, 'orange')
axs[1].text (x = 6, y = 1, s = "SETOSA", fontsize = 10)
axs[1].text (x = 4.3, y = 5, s = "VIRGINICA\n & \nVERSICOLOR", fontsize = 10)

# Plot 3.
ratio = df["sepal_length"]
mean = df["sepal_length"].mean()
    # Scatterplots, grouped by species
for name, group in df.groupby("species"):
    axs[2].scatter(x = group.index, y = ratio[group.index], label=name)
axs[2].set(title = "SEPAL LENGTH")
axs[2].axvspan(0, 50, facecolor='blue', alpha=0.05)
axs[2].axvspan(50,100, facecolor='orange', alpha=0.05)
axs[2].axvspan(100, 150, facecolor='green', alpha=0.05)
axs[2].axhline (mean)
axs[2].text (x = 1, y = mean, s = "Mean\n")
axs[2].legend()

# Plot 4.
ratio = df["petal_length"]
mean = df["petal_length"].mean()
for name, group in df.groupby("species"):
    axs[3].scatter(x = group.index, y = ratio[group.index], label=name)
axs[3].set(title = "PETAL LENGTH")
    # Background colors.
axs[3].axvspan(0, 50, facecolor='blue', alpha=0.05)
axs[3].axvspan(50,100, facecolor='orange', alpha=0.05)
axs[3].axvspan(100, 150, facecolor='green', alpha=0.05)
    # Mean line.
axs[3].axhline (mean)
    # Text.
axs[3].text (x = 1, y = mean, s = "Mean\n")
axs[3].legend()

plt.show()
````

| Plot 3. Setosa has the smallest sepal length, followed by versicolor. Virginica has the largest sepal length. All three flowers can be found with sepal lengths of a very silimar size. |
| Plot 4. Setosa has the smallest petal length and it is much smaller then in the other 2 flowers. Same as the sepal lenght, virginica has the largest petal length. It is easier to distinguish species based on their petal lenght then sepal lebgth. |

Versicolor and virginica values are very similar, they overlap and it would be hard to recognize versicolor flower based only on the sepal length. |
| PETAL LENGTH: Most of the values of petal length are aroun 1.5. There is a gap in the values around 2.5, meaning there are no flowers with petal lenght of this sze. When grouped by species it is visible how setosa has by far the smallest petal length. It is followed by versicolor, and then virginica that has the largers petal length.
